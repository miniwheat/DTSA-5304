```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('data/df.csv')
df['Admission Rate'] = round(df['Admission Rate'], ndigits = 3)
#df.head()
```

```python
single = alt.selection_single(on='mouseover', nearest=False, empty='none')
# single = alt.selection_single()

admits = alt.Chart(df).mark_circle().encode(
            x = 'Applications',
            y = 'Admissions',
            color=alt.Color('Admission Rate', legend=alt.Legend(orient='bottom')),
            tooltip = ['Name', 'Admission Rate']
).add_selection(single).interactive().properties(title = 'US College and University Admissions')

base_scores = alt.Chart(df).mark_bar().encode(
).transform_filter(single)

scores = (base_scores.encode(alt.X('SATVR75', scale = alt.Scale(domain = (0, 800)))) &
          base_scores.encode(alt.X('SATMT75', scale = alt.Scale(domain = (0, 800)))) &
          base_scores.encode(alt.X('ACTEN75', scale = alt.Scale(domain = (0, 36)))) &
          base_scores.encode(alt.X('ACTMT75', scale = alt.Scale(domain = (0, 36))))
         )

#(admits | scores).save('admissions.html', embed_options = {'renderer':'svg'})
admits | scores
```
