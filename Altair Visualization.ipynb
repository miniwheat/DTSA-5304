```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('data/df.csv')
df['Admission Rate'] = round(df['Admission Rate'], ndigits = 3)
df.head()
```

```python
admits = alt.Chart(df).mark_circle().encode(
            x = 'Applications',
            y = 'Admissions',
#             color = 'Region',
            tooltip = ['Name', 'Admission Rate']
).interactive(
).properties(
    title = 'US College and Univ Admissions'
)


admits.save('admissions.html', embed_options = {'renderer':'svg'})

admits
```

```python
interval = alt.selection_interval()

admits = alt.Chart(df).mark_circle().encode(
            x = 'Applications',
            y = 'Admissions',
            tooltip = ['Name', 'Admission Rate']
).add_selection(interval)

test_scores = alt.Chart(df).mark_bar().encode(
    x = 'count()',
    y = 'Region',
).transform_filter(interval)

test = admits | test_scores
# test.save('test.html', embed_options = {'renderer':'svg'})
# alternatively (admits | test_scores).save(...)
test
```

```python
single = alt.selection_single(on='mouseover', nearest=False)

admits = alt.Chart(df).mark_circle().encode(
            x = 'Applications',
            y = 'Admissions',
            tooltip = ['Name', 'Admission Rate']
).add_selection(single)

o1 = alt.Chart(df).mark_bar().encode(
    x = 'SATVR75'
).transform_filter(single)

o2 = alt.Chart(df).mark_bar().encode(
    x = 'SATMT75'
).transform_filter(single)

o3 = alt.Chart(df).mark_bar().encode(
    x = 'ACTEN75'
).transform_filter(single)

o4 = alt.Chart(df).mark_bar().encode(
    x = 'ACTMT75'
).transform_filter(single)

admits | (o1 & o2 & o3 & o4)
```

```python
single = alt.selection_single(on='mouseover', nearest=False, empty='none')
# single = alt.selection_single()

admits = alt.Chart(df).mark_circle().encode(
            x = 'Applications',
            y = 'Admissions',
#             color = 'Admission Rate',
            color=alt.Color('Admission Rate', legend=alt.Legend(orient='bottom')),
            tooltip = ['Name', 'Admission Rate']
).add_selection(single).interactive().properties(title = 'US College and University Admissions')

base_scores = alt.Chart(df).mark_bar().encode(
).transform_filter(single)

scores = (base_scores.encode(alt.X('SATVR75', scale = alt.Scale(domain = (0, 800)))) &
          base_scores.encode(alt.X('SATMT75', scale = alt.Scale(domain = (0, 800)))) &
          base_scores.encode(alt.X('ACTEN75', scale = alt.Scale(domain = (0, 36)))) &
          base_scores.encode(alt.X('ACTMT75', scale = alt.Scale(domain = (0, 36))))
         )

(admits | scores).save('admissions.html', embed_options = {'renderer':'svg'})
admits | scores
```
